```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist={'n_estimators':[int(x) for x in np.linspace(start=10,stop=150,num=15)],
            'max_features':['auto','sqrt'],
            'min_samples_leaf':[1,2,3,4,5,6,7,8,9],
            'max_depth':[int(x) for x in np.linspace(2,30,10)],
            'min_samples_split':[int(x) for x in np.linspace(2,15,5)]}
```

```python
def hyper(rfr,param,X,y):
    random=RandomizedSearchCV(estimator=rfr,param_distributions=param,n_jobs=-1,cv=5,n_iter=20)
    random.fit(X,y)
    ht_params=random.best_params_
    ht_score=random.best_score_
    return ht_params,ht_score
```

```python
parameters,scores=hyper(clf,param_dist,X,y)
```
